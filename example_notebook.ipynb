```python
# example_notebook.ipynb

import pandas as pd
import matplotlib.pyplot as plt
import datetime

# 오늘 날짜
today = datetime.date.today().isoformat()

# 예시 데이터: 주식 가격 (Apple Inc.)
data = {
    'Date': pd.date_range(start='2023-01-01', periods=10, freq='D'),
    'Close': [150, 152, 154, 153, 155, 157, 156, 158, 160, 162]
}

# DataFrame으로 변환
df = pd.DataFrame(data)

# 날짜 필터링 (오늘 날짜 이후 데이터만 사용)
df_filtered = df[df['Date'] <= today]

# 그래프 그리기
plt.figure(figsize=(10, 5))
plt.plot(df_filtered['Date'], df_filtered['Close'], marker='o')
plt.title('Apple Inc. Stock Prices')
plt.xlabel('Date')
plt.ylabel('Close Price')
plt.grid(True)
plt.savefig('/content/stock_prices.png')  # 그래프 저장
plt.show()

print("Notebook executed successfully!")
```
